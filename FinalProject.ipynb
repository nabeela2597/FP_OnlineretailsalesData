```python
import pandas as pd

#loading the retail dataset

df = pd.read_csv('Online-Retail.csv', encoding='unicode_escape')
df
```

```python
df.head()
```

```python
df.info()
```

```python
dfc= df.dropna(subset=['Customer ID'])
dfc
```

```python
dfc['Description'].fillna('Unknown', inplace=True)
dfc
```

```python
#normalize the price column

dfc['Price']= (dfc['Price']- dfc['Price'].mean())/dfc['Price'].std()
dfc
```

```python
dfc = dfc[dfc['Quantity'] > 0]
dfc
```

```python
dfc.info()
```

```python
dfc.describe()
```

```python
country_sales= dfc.groupby('Country')['Quantity'].sum().reset_index()
```

```python
country_sales
```

```python
dfc['TotalSales']= dfc['Quantity']* dfc['Price']
```

```python
country_revenue = dfc.groupby('Country')['TotalSales'].sum().reset_index()
```

```python
country_revenue
```

```python
country_sales.head()
```

```python
country_revenue.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,6))
plt.bar(country_revenue['Country'], country_revenue['TotalSales'], color='brown')
plt.title('Total sales revenue by country')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation= 90)
plt.show()
```

```python
dfc['InvoiceDate']= pd.to_datetime(dfc['InvoiceDate'])
dfc
```

```python
daily_sales = dfc.groupby(dfc['InvoiceDate'].dt.date)['Quantity'].sum().reset_index()
```

```python
daily_sales
```

```python
plt.figure(figsize=(15,6))
plt.plot(daily_sales['InvoiceDate'], daily_sales['Quantity'], marker='o', color='orange')
plt.title('Total Quantity sold over time')
plt.xlabel('Date')
plt.ylabel('Total Quantities sold ')
plt.xticks(rotation= 45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(dfc['Price'], bins=5, color='red', edgecolor='orange')
plt.title('Distribution of Normalized prices')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
```
